```python
import os
from PIL import Image
import torch
from transformers import BitsAndBytesConfig, pipeline
import pandas as pd
from transformers import pipeline, BitsAndBytesConfig
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
image_folder_path = '/content/drive/My Drive/downloads'

quantization_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_compute_dtype=torch.float16
)
```

```python
model_id = "llava-hf/llava-1.5-7b-hf"
pipe = pipeline("image-to-text", model=model_id, model_kwargs={"quantization_config": quantization_config})
data = []

for image_filename in os.listdir(image_folder_path):
    image_path = os.path.join(image_folder_path, image_filename)
    if os.path.isfile(image_path):
        try:
            image = Image.open(image_path)
            prompt = "USER: <image>\nPlease give me a detailed caption of the image\nASSISTANT:"

            outputs = pipe(image, prompt=prompt, generate_kwargs={"max_new_tokens": 200})
            caption = outputs[0]['generated_text']

            print(f"Caption for {image_filename}: {outputs[0]['generated_text']}")
            print(f"Absolute path: {image_path}")
            data.append({'Name': image_filename, 'Caption': caption, 'Path': image_path})

        except Exception as e:
            print(f"Error processing {image_filename}: {e}")

df = pd.DataFrame(data)
parquet_path = '/content/drive/My Drive/image_data.parquet'
directory = os.path.dirname(parquet_path)
os.makedirs(directory, exist_ok=True)
df.to_parquet(parquet_path, index=False)
```
